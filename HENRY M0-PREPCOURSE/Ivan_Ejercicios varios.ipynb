Ramificaciones, ciclos e iteradores

```python
contador = 0
while contador < 10:
    print(contador)
    contador += 1

    if contador > 6:
        break
print("")

#forma 1 de mostrar una lista
frutas = ['manzana', 'pera', 'mango']
iterador = iter(frutas)
print(next(iterador))
print(next(iterador))
print(next(iterador))
print("")

#forma 2 de mostrar una lista
frutas = ['manzana', 'pera', 'mango']
for fruta in frutas:
    print(fruta)
```

```python
estudiantes = {
    'mexico': 10,
    'colombia': 15,
    'puerto_rico': 4,
}

for pais in estudiantes:
    print(pais)
print("")

for pais in estudiantes.keys():
    print(pais)
print("")

for numero_de_estudiantes in estudiantes.values():
    print(pais)
print("")

for pais, numero_de_estudiantes in estudiantes.items():
    print(pais)
print("")


"""
Podemos modificar el comportamiento de un bucle for mediante los keywords break y continue.
break termina el bucle y permite continuar con el resto del flujo de nuestro programa.
continue termina la iteración en curso y continua con el siguiente ciclo de iteración.
"""
```

```python
for n in range(1,10):
    print(n)
print("")

n = 1
while (n < 10):
    print(n)
    n += 1 #es lo mismo que decir n = n + 1

print("")
contador = 0
while contador < 10:
    print(contador)
    contador += 1       # contador = contador + 1

    if contador > 6:    # Cuando contador sea mayor que 6 terminara la iteración.
        break
```
