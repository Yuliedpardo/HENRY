```python
def NumeroBinario(numero, base = 2):
    '''
    Esta función recibe como parámetro un número entero mayor ó igual a cero y lo devuelve en su 
    representación binaria. Debe recibir y devolver un valor de tipo entero.
    En caso de que el parámetro no sea de tipo entero y mayor a -1 retorna nulo.
    '''
    
    if (type(numero) != int) or (numero < 0):            # validación dato de entrada
        return None
    
    resul = ''
    invert = ''
    origen = numero

    while numero >= base:                               # continuar mientras que el numero a dividir sea mayor a 2
        resul += str(numero % base)                     # convertir a str y concatenar los residuos
        numero //= base                                 # divide el numero entre 2 (muestra la parte entera)
    resul += str(numero % base)                         # sale del ciclo cuando el numero a dividir es menor a 2, aun asi intenta la división y almacena el residuo

    for i in resul[::-1]:                               # repasa la cadena pero en reversa -1 y lo concatena en una nueva variable para mostrar el resultado invertido
        invert += i             

    print (origen, " ---> ", invert)                    # otra manera es: numero[::-1]

    return int(invert)
```

```python
def NumeroDecimal (binario, base=2):
    '''
    convierte numero binario a decimal
    '''
    decimal = 0

    for i,j in enumerate (str(binario)[::-1]):      # toma el numero binario, lo convierte a srt y lo empieza a recorrer al revés
        decimal += (base**i)*int(j)                 # eleva 2 a la potencia según su posición y lo multiplica por cada posicion binaria convertida a entero
        # decimal = decimal * (base**i) * int(j)
    print (binario, " ---> ", decimal,"\n")

    return
```
